```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier



# Preprocess the data and train the model (same steps as before)

# Assuming df1 is the DataFrame containing the crime data
# Define a threshold for total IPC crimes to classify a district as safe or not
threshold = 1000
df1['Safe'] = df1['TOTAL IPC CRIMES'].apply(lambda x: 1 if x < threshold else 0)

# Features and target variable
X = df1[['MURDER', 'ATTEMPT TO MURDER', 'RAPE', 'KIDNAPPING & ABDUCTION', 
         'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS', 'DACOITY', 'ROBBERY', 
         'THEFT', 'AUTO THEFT', 'RIOTS', 'CHEATING', 'COUNTERFIETING', 'TOTAL IPC CRIMES']]
y = df1['Safe']

# Train the model
model = RandomForestClassifier()
model.fit(X, y)

# User input for district and year
district = input("Enter the district: ")
year = int(input("Enter the year: "))

# Filter the dataset for the given district and year
district_data = df1[(df1['DISTRICT'] == district) & (df1['YEAR'] == year)][['MURDER', 'ATTEMPT TO MURDER', 'RAPE', 
                                                                          'KIDNAPPING & ABDUCTION', 
                                                                          'KIDNAPPING AND ABDUCTION OF WOMEN AND GIRLS', 
                                                                          'DACOITY', 'ROBBERY', 'THEFT', 
                                                                          'AUTO THEFT', 'RIOTS', 'CHEATING', 
                                                                          'COUNTERFIETING', 'TOTAL IPC CRIMES']]

# Check if data for the given district and year exists
if len(district_data) == 0:
    print("Data not found for the given district and year.")
else:
    # Predict whether the district is safe or not
    prediction = model.predict(district_data)

    # Display the prediction
    if prediction[0] == 0:
        print(f"The district '{district}' in the year {year} is predicted to be unsafe.")
    else:
        print(f"The district '{district}' in the year {year} is predicted to be safe.")
```
